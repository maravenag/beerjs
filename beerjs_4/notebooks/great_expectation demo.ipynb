### Cargar el dataset original y dividir en train/test

```python
from random import randint
from random import choice

def random_instance():
    return {
    "age": randint(60, 80),
    "sex": choice(['male','female']),
    "bmi": randint(25, 35),
    "children": randint(0,2),
    "smoker":choice(["yes","no"]),
    "region":choice(['northeast', 'southwest', 'southeast', 'northwest']),
    "charges":randint(37931,50575)
}
```

```python
new_dist = []
for x in range(0,300):
    new_dist.append(random_instance())
```

```python
test = test.append(pd.DataFrame(new_dist),ignore_index=True, sort=True)
```

```python
test.hist()
```
